```python
# json파일 불러오기
import json


path = r"/Users/jeong-uihyeon/chatbot/file/Sample/02.라벨링데이터/국정감사/16/LAB_16대_2000_2000년10월20일_국정감사_교육위원회_0001(030043).json" # 임시 경로

def load_json(file_path):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file) # 전체 내용
    except FileNotFoundError:
        data = "파일을 찾을 수 없습니다."
    except json.JSONDecodeError:
        data = "유효하지 않은 JSON 형식입니다."
    return data

json_data = load_json(path)
print(type(json_data['question']["comment"]))
print(json_data['answer']['comment'])
```

```python
# 이제 모든 경로를 탐색해서 json 파일로부터 질문 답변을 받아오기
import os
import json

def read_json_files(directory):
    question = []
    answer = []
    
    # 디렉토리와 하위 디렉토리를 탐색
    for root, dirs, files in os.walk(directory):
        for file in files:
            # 파일 확장자가 .json인지 확인
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                # JSON 파일 읽기
                data = load_json(file_path)
                # json파일에서 question / answer 쌍으로 추출
                question.append(data['question']["comment"])
                answer.append(data['answer']["comment"])
                
    return question, answer


directory_path = r"/Users/jeong-uihyeon/chatbot/file/Sample/02.라벨링데이터/국정감사/16"  #원하는 경로 (일단 한정해 두었음)
questions, answers = read_json_files(directory_path)
print(questions)
print(len(questions))
print(answers)
```

```python
# 입력 받은 데이터를 임베딩 (sentence bert 사용)
# 3.12 버전에서는 호환이 안되니 3.11 이하 버전 사용 (여기선 3.9.6이용)
from sentence_transformers import SentenceTransformer


def embed_text(texts, model):
    return model.encode(texts, show_progress_bar=True)

# Sentence-BERT 모델 로드
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# 결합된 텍스트를 벡터로 변환 (지금은 질문 하나만 변환해서 확인)
combined_vectors = embed_text(questions[0], model)

print(combined_vectors)
```

```python
import faiss
import numpy as np

# 질문 전부를 임베딩하기
embeddings = model.encode(questions)

#벡터의 차원 확인
vector_dimension = len(embeddings[0])

#Faiss index 생성
##faiss index는 효율적인 정보 검색을 도와줌 (여기서는 IndexFlatL2:가장 기본적인 인덱스 를 이용해볼 예정)
index = faiss.IndexFlatL2(vector_dimension)

# Faiss는 float32 타입을 요구하기 때문에 float32로 변환
embeddings_np = np.array(embeddings).astype('float32')

# 인덱스에 벡터 추가
index.add(embeddings_np)
```

```python
# 저장된 문장 검색하는 방법

# 검색할 쿼리 문장
query = "그 샘플을 보여 주십시오.  그다음에 노벨상 수상에 대한 것이라든지 남북정상회담에 대한 것 또 대통령의 대북 햇볕정책에 대해서 중국 전체 주민들을 대상으로 하지는 않는다 하더라도 북경 주민들을 대상으로 해서 공감대에 대한 여론조사같은 것을 한 번 해보신 적이 있습니까?"

# 쿼리 문장을 벡터로 변환
query_vector = model.encode([query])

# 가장 유사한 5개의 벡터 검색
k = 5
distances, indices = index.search(query_vector, k)

# 결과 출력
for i in range(k):
    print(f"유사도: {distances[0][i]}, 인덱스: {indices[0][i]}, 문장: {questions[indices[0][i]]}")
```
